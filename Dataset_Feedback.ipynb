```python
from googletrans import Translator
translator = Translator()

def translate_column_name(column_name):
    translated = translator.translate(column_name, src='en', dest='id')
    return translated.text


df.columns = [translate_column_name(column) for column in df.columns]
df.T
```

```python
# Mempersingkat nama kolom
column_mapping = {
    'Berdasarkan hasil pembelajaran yang ditentukan, seberapa puaskah Anda dengan pengalaman pelatihan ini?': 'Kepuasan terhadap pelatihan',
    'Saya akan merekomendasikan pengalaman hari ini kepada para profesional pendidikan lainnya.': 'Merekomendasikan kepada profesional pendidikan',
    'Menghadiri pelatihan ini meningkatkan pemahaman saya tentang bagaimana alat Microsoft dapat digunakan untuk meningkatkan hasil siswa.' : 'Meningkatkan pemahaman tentang alat microsoft ',
    'Saya lebih cenderung merekomendasikan Microsoft sebagai hasil dari berpartisipasi dalam pelatihan ini.' : 'Merekomendasikan Microsoft',
    'Presenter memiliki pengetahuan dan pemahaman yang menyeluruh tentang konten yang disampaikan selama sesi.Rinaldi' : 'Pengetahuan dan pengalaman presenter',
    'Konten disajikan dengan jelas, ringkas, dan dengan cara yang menarik dan interaktif.Rinaldi' : 'Konten yang disajikan',
    'Laju sesi itu tepat.Rinaldi' : 'Ketepatan sesi',
    'Harap berikan umpan balik tambahan yang ingin Anda bagikan tentang pengalaman Anda dengan pelatih yang menyampaikan sesi Anda hari ini.' : 'Umpan balik terhadap pelatih',
    'Tema' : 'Topik'
}

df = df.rename(columns=column_mapping)
df.T
```

```python
# Mencari modus kata-kata dalam kolom
# Menghitung frekuensi kemunculan kata-kata
frekuensi = df['Umpan balik terhadap pelatih'].value_counts()
frekuensi
```

Diketahui bahwa modus pertama adalah "-" maka kita menggunakan modus keduanya, yaitu "Sudah bagus".

```python
# Mengisi kolom yang kosong

# Mengisi kolom yang kosong hanya jika kosong
modus_kedua = 'Sudah bagus'
df['Umpan balik terhadap pelatih'] = df['Umpan balik terhadap pelatih'].apply(lambda x: modus_kedua if pd.isna(x) else x)

# Pada modus pertama yang banyak muncul adalah "-", maka kita ganti isinya menjadi modus kedua
# Mengganti nilai pada kolom yang mengandung strip ("-") dengan modus kedua
df['Umpan balik terhadap pelatih'] = df['Umpan balik terhadap pelatih'].apply(lambda x: modus_kedua if pd.notna(x) and '-' in x else x)
df.T
```

```python
# mengkonversi huruf ke angka
def convert_kepuasan_ke_angka(kepuasan) :
    if kepuasan == "Sangat Puas":
        return 5
    elif kepuasan == "Puas":
        return 3
    else:
        return 0

kolom_yang_diubah = ['Kepuasaan terhadap pelatihan', 'Meningkatkan pemahaman untuk produktivitas kerja', 'Pengetahuan dan pemahaman pemateri', 'Konten yang disajikan']

# Mengubah nilai langsung pada setiap kolom
for kolom in kolom_yang_diubah:
    df_3[kolom] = df_3[kolom].apply(convert_kepuasan_ke_angka)

df_3
```

```python
# Mengecek data yang missing
df_5.isna().sum()
```

```python
# Normalisasi kolom 'Sumber informasi'
def normalize_source_info(info):
    # Menyatukan semua informasi menjadi list
    info_list = info.split(';')

    # Menghapus elemen yang kosong dari list
    info_list = [word.strip() for word in info_list if word.strip()]

    # Menggabungkan kembali list menjadi string ternormalisasi
    normalized_info = ';'.join(info_list)

    return normalized_info

# Mengaplikasikan fungsi normalisasi ke kolom 'sumber_informasi'
df_5['Sumber informasi'] = df_5['Sumber informasi'].apply(normalize_source_info)
df_5
```
